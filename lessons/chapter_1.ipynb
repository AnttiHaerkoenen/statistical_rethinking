```python
import numpy as np
import pandas as pd
from scipy.stats import binom
import matplotlib.pyplot as plt
```

## 2M1.

```python
prior = np.repeat(1, 101)
data = {
    'case_1': np.array([3, 0]),
    'case_2': np.array([3, 1]),
    'case_3': np.array([5, 2]),
}
grid = np.linspace(0, 1, 101)

def get_posterior(data, grid, prior):
    likelihood = binom.pmf(data[0], data.sum(), grid)
    unscaled_posterior = likelihood * prior
    posterior = unscaled_posterior / unscaled_posterior.sum()
    return posterior
```

### (1)

```python
posterior = get_posterior(data['case_1'], grid, prior)
plt.plot(grid, posterior)
plt.show()
```

### (2)

```python
posterior = get_posterior(data['case_2'], grid, prior)
plt.plot(grid, posterior)
plt.show()
```

### (3)

```python
posterior = get_posterior(data['case_3'], grid, prior)
plt.plot(grid, posterior)
plt.show()
```
